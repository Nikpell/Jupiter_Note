Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
    plt.figure(figsize=(16, 10))

sns.set(font_scale=1.4)

sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm')

plt.title('Correlation matrix');
```

Из корреляционной матрицы следует, что цена коррелирует с жилой площадью и уровнем конструкции и дизайна(наиболее сильно),жилой площадью выше земли и жилой площадью у соседей. 

```python
sns.jointplot(x=df['price'], y=df['sqft_living'], kind='reg')
plt.xlabel('Цена')
plt.ylabel('Жилая площадь')
plt.xticks(rotation=30);
```

Прямая корреляция между ценой и жилой площадью. Цена дома зависит от жилой площади. Чем больше площадь- тем выше цена

```python
sns.jointplot(x=df['price'], y=df['grade'], kind='reg')
plt.xlabel('Цена')
plt.ylabel('Уровнь конструкции и дизайна')
plt.xticks(rotation=30);
```

Прямая корреляция между ценой и уровнем конструкции и дизайна. Цена дома зависит от уровня конструкции и дизайна. Чем выше уровнь конструкции и дизайна- тем выше цена

```python
sns.jointplot(x=df['price'], y=df['sqft_above'], kind='reg')
plt.xlabel('Цена')
plt.ylabel('Жилая площадь выше земли')
plt.xticks(rotation=30);
```

Прямая корреляция между ценой и жилой площадью выше земли. Цена дома зависит от жилой площади выше земли. Чем больше жилая площадью выше земли- тем выше цена
